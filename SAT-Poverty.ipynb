```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('scores.csv')
df=df[['School ID', 'School Name', 'Borough','City','Average Score (SAT Math)', 'Average Score (SAT Reading)','Average Score (SAT Writing)']]
df=df.dropna()
df['Average Score (Math+Reading)']=df['Average Score (SAT Math)']+df['Average Score (SAT Reading)']
df=df.rename(columns={'School ID': 'DBN'})

df2 = pd.read_excel('Demographic_Snapshot_2017-18_to_2021-22__Public_.xlsx')
df2=df2[['DBN','% Poverty']]
df2=df2.replace('Above 95%', 0.95)
df2=df2.replace('Below 5%', 0.05)
df2=df2.groupby('DBN').mean()

df3=pd.merge(df, df2, how='inner', left_on='DBN', right_on='DBN')

df3['% Poverty']=df3['% Poverty']*100

sns.jointplot(x=df3['% Poverty'], y=df3['Average Score (Math+Reading)'], edgecolor="white", data=df3, color="royalblue")
plt.suptitle('NYC Public School SAT Scores', y=1)
plt.xlabel('% of Students in Poverty')
plt.ylabel('Average SAT Score (Math + Reading)')
plt.savefig("NYC Public School SAT Scores 2.png", dpi=150)
```
